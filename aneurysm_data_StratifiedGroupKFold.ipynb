#### Podział na zbiory test i train z wykorzystaniem StratifiedGroupKFold

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedGroupKFold

df = pd.read_csv('/content/aneurysm_concatted_cleaned.csv')

print(f"Kształt zbioru danych: {df.shape}")
```

```python
print(df.columns)
```

```python
target_column = 'label'
group_column = 'patient_id'

X = df.drop(columns=[target_column])
y = df[target_column]
groups = df[group_column]
```

```python
# Konfiguracja StratifiedGroupKFold

n_splits = 5
sgkf = StratifiedGroupKFold(n_splits=n_splits)

print(f"--- ROZPOCZĘCIE PODZIAŁU NA {n_splits} FOLDÓW ---\n")

# iteracja po foldach i weryfikacja podziału
for fold, (train_idx, test_idx) in enumerate(sgkf.split(X, y, groups)):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    groups_train, groups_test = groups.iloc[train_idx], groups.iloc[test_idx]

    # Sprawdzenie proporcji klas (0 vs 1) w train i test
    train_dist = y_train.value_counts(normalize=True).to_dict()
    test_dist = y_test.value_counts(normalize=True).to_dict()

    # Sprawdzenie, czy grupy (ID) się nie pokrywają (wynikiem powinien być pusty set)
    overlapping_groups = set(groups_train).intersection(set(groups_test))

    print(f"Fold {fold + 1}:")
    print(f"  Rozmiar: Train = {len(train_idx)}, Test = {len(test_idx)}")
    print(f"  Proporcje klas Train: { {k: f'{v:.2%}' for k, v in train_dist.items()} }")
    print(f"  Proporcje klas Test : { {k: f'{v:.2%}' for k, v in test_dist.items()} }")
    print(f"  Wspólne ID między Train a Test: {len(overlapping_groups)} (Oczekiwane: 0)")
    print("-" * 50)
```
